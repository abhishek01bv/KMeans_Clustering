```python
df.drop('CustomerID',axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Genre'] = le.fit_transform(df['Genre'])
df.value_counts('Genre')
#where 0 is Female and 1 is Male
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
#there is no correlations between the data
```

```python
plt.scatter(df['Annual Income (k$)'], df.Age)
```

```python
from sklearn.cluster import KMeans

X = df.iloc[:,1:]
X.head()
```

```python
model = KMeans(n_clusters=2, random_state=10)
model.fit(X)
```

```python
labels= model.labels_
labels
```

```python
model.cluster_centers_
```

```python

plt.scatter(X['Age'], X['Annual Income (k$)'], c=labels, cmap='viridis')
```

```python
plt.scatter(X['Age'], X['Annual Income (k$)'], c=labels, cmap='viridis')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title('K-Means Clustering')
plt.show()
```
